```python
# Extract all words from columns 1 to 22, excluding missing values
all_words = df.iloc[:, 0:24].stack().dropna()

# Create a new DataFrame with all words
df_words_only = pd.DataFrame({'all_words': all_words})
```

```python
df_words_only
```

```python
df_unique = df_words_only['all_words'].unique()
```

```python
# Count the occurrences of each unique word
word_counts = df_words_only['all_words'].value_counts()

df_unique_counts = pd.DataFrame({'unique_words': word_counts.index, 'count': word_counts.values})
```
